```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
import seaborn as sns

%matplotlib inline
```

# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_path = Path("Resources/whale_returns.csv")
whale_df = pd.read_csv(whale_path, infer_datetime_format=True, parse_dates=True)
whale_df['Date'] = pd.to_datetime(whale_df["Date"], dayfirst = True)
whale_df.set_index(whale_df['Date'], inplace=True)
whale_df.drop(columns=['Date'],inplace=True)
whale_df.head()
```

```python
# Count nulls
whale_df.isnull().sum()
```

```python
# Drop nulls
whale_df = whale_df.dropna()
whale_df.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_path = Path("Resources/algo_returns.csv")
algo_df = pd.read_csv(algo_path, infer_datetime_format=True, parse_dates=True)
algo_df['Date'] = pd.to_datetime(algo_df["Date"], dayfirst = True)
algo_df.set_index(algo_df['Date'], inplace=True)
algo_df.drop(columns=['Date'],inplace=True)
algo_df.head()
```

```python
# Count nulls
algo_df.isnull().sum()
```

```python
# Drop nulls
algo_df = algo_df.dropna()
algo_df.isnull().sum()
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp_path = Path("Resources/sp500_history.csv")
sp_df = pd.read_csv(sp_path, infer_datetime_format=True, parse_dates=True)
sp_df['Date'] = pd.to_datetime(sp_df["Date"], dayfirst = True)
sp_df.set_index(sp_df['Date'], inplace=True)
sp_df.drop(columns=['Date'],inplace=True)
sp_df.head()
```

```python
# Check Data Types
sp_df.dtypes
```

```python
# Fix Data Types
sp_df['Close'] = sp_df['Close'].str.replace('$', '')
sp_df['Close'] = sp_df['Close'].astype('float')
sp_df.dtypes
```

```python
# Calculate Daily Returns
sp_df = sp_df.pct_change()
sp_df.head()
```

```python
# Drop nulls
sp_df = sp_df.dropna()
sp_df.isnull().sum()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp_df.columns = ["S&P 500"]
sp_df.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
all_df = pd.concat([whale_df,algo_df, sp_df], axis='columns', join='inner')
all_df.head()
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for the S&P 500 using a 21-day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
3. Choose one portfolio, then calculate and plot the 60-day rolling beta between it and the S&P 500

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
rollingstd_all_df = all_df.rolling(window=21).std()

# Plot the rolling standard deviation
rollingstd_all_df.plot(figsize=(20,10))
```

### Calculate and plot the correlation

```python
# Calculate the correlation
corr_all_df = all_df.corr()

# Display de correlation matrix
corr_all_df
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
algo1_covar = all_df['Algo 1'].rolling(window=21).cov(all_df['S&P 500'])
algo1_covar

# Calculate variance of S&P 500
sp_var = all_df['S&P 500'].rolling(window=21).var()

# Computing beta
algo1_beta = algo1_covar / sp_var

# Plot beta trend
algo1_beta.plot(figsize=(20,10))
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
ewm_all_df = all_df.ewm(halflife=21, adjust=True).std()
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_df = (all_df.mean()*252) / (all_df.std()*(252**0.5))
sharpe_df
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_df.plot(kind='bar')
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

For this demo solution, we fetch data from three companies listes in the S&P 500 index.

* `GOOG` - [Google, LLC](https://en.wikipedia.org/wiki/Google)

* `AAPL` - [Apple Inc.](https://en.wikipedia.org/wiki/Apple_Inc.)

* `COST` - [Costco Wholesale Corporation](https://en.wikipedia.org/wiki/Costco)

```python
# Reading data from 1st stock
goog_path = Path("Resources/goog_historical.csv")
goog_df = pd.read_csv(goog_path, infer_datetime_format=True, parse_dates=True)
goog_df['Trade DATE'] = pd.to_datetime(goog_df['Trade DATE'], dayfirst = True)
goog_df.set_index(goog_df['Trade DATE'], inplace=True)
goog_df.drop(columns=['Trade DATE', 'Symbol'],inplace=True)
goog_df.columns = ['GOOG']
goog_df.head()
```

```python
# Reading data from 2nd stock
aapl_path = Path("Resources/aapl_historical.csv")
aapl_df = pd.read_csv(goog_path, infer_datetime_format=True, parse_dates=True)
aapl_df['Trade DATE'] = pd.to_datetime(aapl_df['Trade DATE'], dayfirst = True)
aapl_df.set_index(aapl_df['Trade DATE'], inplace=True)
aapl_df.drop(columns=['Trade DATE', 'Symbol'],inplace=True)
aapl_df.columns = ['AAPL']
aapl_df.head()
```

```python
# Reading data from 3rd stock
cost_path = Path("Resources/aapl_historical.csv")
cost_df = pd.read_csv(cost_path, infer_datetime_format=True, parse_dates=True)
cost_df['Trade DATE'] = pd.to_datetime(cost_df['Trade DATE'], dayfirst = True)
cost_df.set_index(cost_df['Trade DATE'], inplace=True)
cost_df.drop(columns=['Trade DATE', 'Symbol'],inplace=True)
cost_df.columns = ['COST']
cost_df.head()
```

```python
# Combine all stocks in a single DataFrame
port_df = pd.concat([goog_df,aapl_df,cost_df], axis="columns", join="inner")
port_df.head()
```

```python
# Reset Date index
port_df.sort_index(inplace=True)
```

```python
# Reorganize portfolio data by having a column per symbol
port_df.head()
```

```python
# Calculate daily returns
port_df = port_df.pct_change()

# Drop NAs
port_df.dropna(inplace=True)

# Display sample data
port_df.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]

# Calculate portfolio return
weighted_port_df = port_df.dot(weights)

# Display sample data
weighted_port_df.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
all_port_df = pd.concat([all_df, weighted_port_df], axis='columns',join='inner')
all_port_df.columns = ['SOROS FUND MANAGEMENT LLC', 'PAULSON & CO.INC. ', 'TIGER GLOBAL MANAGEMENT LLC', 'BERKSHIRE HATHAWAY INC', 'Algo 1', 'Algo 2', 'S&P 500', 'Weighted Portfolio']
all_port_df.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
all_port_df.dropna(inplace=True)
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized `std`
all_port_df.std()
```

### Calculate and plot rolling `std` with 21-day window

```python
# Calculate rolling standard deviation
rollingstd_all_port_df = all_port_df.rolling(window=21).std()

# Plot rolling standard deviation
rollingstd_all_port_df.plot(figsize=(20,10))
```

### Calculate and plot the correlation

```python
# Calculate and plot the correlation
corr_all_port_df = all_port_df.corr()
sns.heatmap(corr_all_port_df,vmin=-2, vmax=1)
```

### Calculate and Plot Rolling 60-day Beta for Your Portfolio compared to the S&P 500

```python
# Calculate and plot Beta
cov_port = all_port_df['Weighted Portfolio'].rolling(window=60).cov(all_port_df['S&P 500'])
var_port = all_port_df['S&P 500'].rolling(window=60).var()
beta_port = cov_port / var_port
beta_port.plot(figsize=(20,10))
```

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Calculate Annualzied Sharpe Ratios
sharpe_all_port_df = (all_port_df.mean()*252)/(all_port_df.std()*(252**0.5))
sharpe_all_port_df
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_all_port_df.plot(kind="bar")
```
